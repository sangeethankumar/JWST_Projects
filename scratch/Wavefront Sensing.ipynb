```python
import os
from astropy.io import fits
import matplotlib.pyplot as plt
import numpy as np
```

```python
path_to_fits = '../data/wss'  
fits_files = [f for f in os.listdir(path_to_fits) if f.endswith('.fits')]
fits_data_list = []
for file in fits_files:
    with fits.open(os.path.join(path_to_fits, file)) as hdul:
        fits_data_list.append(hdul[1].data)

len(fits_data_list)
```

```python
plt.imshow(fits_data_list[0], cmap='viridis', vmin=-0.2, vmax=0.5)
plt.colorbar()
```

```python
plt.imshow(fits_data_list[1], cmap='viridis', vmin=-0.2, vmax=0.5)
plt.colorbar()
```

```python
plt.imshow(fits_data_list[-1], cmap='viridis', vmin=-0.2, vmax=0.5)
plt.colorbar()
```
